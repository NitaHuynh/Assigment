## Exploratory Data Analysis


First, let's read the SpaceX dataset into a Pandas dataframe and print its summary


```python
from js import fetch
import io

URL = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_2.csv"
resp = await fetch(URL)
dataset_part_2_csv = io.BytesIO((await resp.arrayBuffer()).to_py())
df=pd.read_csv(dataset_part_2_csv)
df.head(5)
```

First, let's try to see how the `FlightNumber` (indicating the continuous launch attempts.) and `Payload` variables would affect the launch outcome.

We can plot out the <code>FlightNumber</code> vs. <code>PayloadMass</code>and overlay the outcome of the launch. We see that as the flight number increases, the first stage is more likely to land successfully. The payload mass also appears to be a factor; even with more massive payloads, the first stage often returns successfully.


Đầu tiên, hãy thử xem các biến **FlightNumber** (số lần phóng liên tục) và **Payload** (khối lượng hàng hóa) sẽ ảnh hưởng đến kết quả hạ cánh như thế nào.  

Chúng ta có thể vẽ biểu đồ **FlightNumber vs PayloadMass** và phủ thêm kết quả của vụ phóng. Quan sát:  

- Khi **FlightNumber** tăng, khả năng hạ cánh thành công của tàu tăng một cách rõ rệt.  
- **PayloadMass** cũng là một yếu tố: ngay cả với tải trọng lớn, tầng một vẫn thường hạ cánh thành công.


```python
sns.catplot(y="PayloadMass", x="FlightNumber", hue="Class", data=df, aspect = 5)
plt.xlabel("Flight Number",fontsize=20)
plt.ylabel("Pay load Mass (kg)",fontsize=20)
plt.show()
```

🔎 Tham số `aspect` trong `sns.catplot()` là **tỉ lệ chiều rộng trên chiều cao (width/height ratio)** của mỗi subplot.  

- Mặc định: `aspect=1` → chiều rộng và chiều cao bằng nhau (hình vuông).  
- Nếu `aspect > 1` → đồ thị sẽ kéo dài theo chiều ngang.  
- Nếu `aspect < 1` → đồ thị sẽ dẹt lại theo chiều dọc.  

📌 Trong ví dụ bạn để `aspect=5`: nghĩa là **chiều rộng gấp 5 lần chiều cao**, giúp trải dài dữ liệu khi trục X có nhiều giá trị (ở đây là số lần phóng `FlightNumber`).  

👉 Nói ngắn gọn: `aspect` giúp điều chỉnh **tỉ lệ hiển thị của biểu đồ** để dễ quan sát hơn.


Next, let's drill down to each site visualize its detailed launch records.


### TASK 1: Visualize the relationship between Flight Number and Launch Site


Use the function <code>catplot</code> to plot <code>FlightNumber</code> vs <code>LaunchSite</code>, set the  parameter <code>x</code>  parameter to <code>FlightNumber</code>,set the  <code>y</code> to <code>Launch Site</code> and set the parameter <code>hue</code> to <code>'class'</code>


```python
# Vẽ scatter plot: FlightNumber vs LaunchSite
sns.catplot(x="FlightNumber", y="LaunchSite", hue="Class", data=df, aspect=5)
plt.xlabel("Flight Number", fontsize=14)
plt.ylabel("Launch Site", fontsize=14)
plt.title("Relationship between Flight Number and Launch Site", fontsize=16)
plt.show()
```

## 2. Observed Patterns

### 1. Success trend over time
- At all three launch sites, orange dots appear more frequently at higher Flight Numbers.  
- This shows that as the number of launches increases, the success rate also improves → SpaceX learns and progresses over time.  

### 2. Differences across Launch Sites
- **CCAFS SLC 40**: contains both failures (blue) and successes (orange), but later on most launches are successful.  
- **KSC LC 39A**: shows a higher success rate, with many orange points even at relatively early launches.  
- **VAFB SLC 4E**: has fewer data points, but still includes both failures and successes → the trend is less clear compared to the other two sites.  

### 3. Learning from early failures
- In the early stage (low Flight Numbers, ~<20), most points are blue (failures).  
- After ~Flight Number 40 onwards, most points turn orange → clear evidence of technological improvement.  

---

## 3. Conclusion
- Success is **not entirely dependent on the Launch Site**, but mainly related to the number of launches (**accumulated experience**).  
- However, some sites such as **KSC LC 39A** appear to provide more favorable conditions → the success rate seems higher compared to **VAFB SLC 4E**.


## Patterns quan sát được

### 1. Xu hướng thành công tăng theo thời gian
- Ở cả 3 bãi phóng, các chấm màu cam xuất hiện nhiều hơn ở các Flight Number lớn.  
- Điều này cho thấy: khi số lần phóng tăng, tỉ lệ thành công cũng tăng → SpaceX học hỏi, cải tiến qua thời gian.  

### 2. Khác biệt giữa các Launch Site
- **CCAFS SLC 40**: có cả thất bại (xanh) và thành công (cam), nhưng về sau đa phần là thành công.  
- **KSC LC 39A**: tỉ lệ thành công cao hơn, nhiều điểm cam ngay từ những lần phóng không quá muộn.  
- **VAFB SLC 4E**: ít dữ liệu hơn, nhưng vẫn có cả thất bại lẫn thành công → chưa rõ xu hướng mạnh bằng 2 site kia.  

### 3. Học hỏi từ thất bại ban đầu
- Ở giai đoạn đầu (Flight Number nhỏ, ~<20), phần lớn là điểm xanh (thất bại).  
- Sau ~Flight Number 40 trở đi, đa số chuyển sang màu cam → minh chứng rõ cho tiến bộ công nghệ.  

---

## 3. Kết luận rút ra
- Thành công **không phụ thuộc hoàn toàn vào Launch Site**, mà chủ yếu liên quan đến số lần phóng (**kinh nghiệm tích lũy**).  
- Tuy nhiên, một số bãi phóng như **KSC LC 39A** dường như có điều kiện thuận lợi hơn → tỉ lệ thành công nhìn hơn so với **VAFB SLC 4E**.


---
Trong **Seaborn** (hàm `catplot`, `scatterplot`, ...), tham số `hue` dùng để phân biệt dữ liệu bằng **màu sắc** dựa trên giá trị của một cột trong DataFrame.  

Cụ thể trong bài này:  
- Cột `Class` = kết quả hạ cánh (landing outcome).  
  - `Class = 1`: hạ cánh **thành công**  
  - `Class = 0`: hạ cánh **thất bại**  

👉 Khi thêm `hue="Class"`:  
- Các điểm dữ liệu trên biểu đồ sẽ được **tô màu khác nhau** tùy thuộc vào giá trị `Class`.  
- Nhờ đó, bạn có thể nhìn trực quan xem **các lần phóng (FlightNumber)** hoặc **khối lượng tải trọng (PayloadMass)** ảnh hưởng thế nào đến **khả năng hạ cánh thành công/thất bại**.  

Ví dụ:  
- Nếu **không có `hue`** → tất cả điểm cùng một màu → khó phân biệt thành công/thất bại.  
- Nếu **có `hue="Class"`** → điểm màu xanh = thành công, điểm màu đỏ = thất bại → dễ quan sát xu hướng.

### TASK  3: Visualize the relationship between success rate of each orbit type


Next, we want to visually check if there are any relationship between success rate and orbit type.


Let's create a `bar chart` for the sucess rate of each orbit


```python
# Tính tỉ lệ thành công trung bình theo Orbit
orbit_success = df.groupby("Orbit")["Class"].mean()

# Vẽ bar chart
orbit_success.plot(kind='bar', figsize=(10,6))
plt.xlabel("Orbit Type", fontsize=14)
plt.ylabel("Success Rate", fontsize=14)
plt.title("Success rate of each orbit type", fontsize=16)
plt.show()
```

Analyze the plotted bar chart to identify which orbits have the highest success rates.


The orbits with the highest success rate are ES-L1, GEO, HEO, and SSO, all with 100% success. Meanwhile, the lowest success rate is observed in SO (0%) and GTO (~50%).

### TASK  4: Visualize the relationship between FlightNumber and Orbit type


For each orbit, we want to see if there is any relationship between FlightNumber and Orbit type.


```python
# Vẽ scatter plot: FlightNumber vs Orbit type
sns.catplot(x="FlightNumber", y="Orbit", hue="Class", data=df, aspect=5)
plt.xlabel("Flight Number", fontsize=14)
plt.ylabel("Orbit Type", fontsize=14)
plt.title("Relationship between FlightNumber and Orbit type", fontsize=16)
plt.show()
```

### TASK  6: Visualize the launch success yearly trend


You can plot a line chart with x axis to be <code>Year</code> and y axis to be average success rate, to get the average launch success trend.


The function will help you get the year from the date:


```python
# Hàm trích xuất năm từ cột Date
year = []
def Extract_year():
    for i in df["Date"]:
        year.append(i.split("-")[0])
    return year

Extract_year()
df['Date'] = year
df.head()
```

```python
# Nhóm theo năm và tính tỉ lệ thành công trung bình
yearly_success = df.groupby("Date")["Class"].mean().reset_index()

# Vẽ line chart
plt.figure(figsize=(10,6))
plt.plot(yearly_success["Date"], yearly_success["Class"], marker='o')

plt.xlabel("Year", fontsize=14)
plt.ylabel("Success Rate", fontsize=14)
plt.title("Trend of Success Rate (by Year)", fontsize=16)
plt.grid(True)
plt.show()
```

you can observe that the sucess rate since 2013 kept increasing till 2020


## Features Engineering


By now, you should obtain some preliminary insights about how each important variable would affect the success rate, we will select the features that will be used in success prediction in the future module.


```python
features = df[['FlightNumber', 'PayloadMass', 'Orbit', 
               'LaunchSite', 'Flights', 'GridFins', 'Reused', 
               'Legs', 'LandingPad', 'Block', 'ReusedCount', 
               'Serial']]
features.head()
```

### TASK  7: Create dummy variables to categorical columns


Use the function <code>get_dummies</code> and <code>features</code> dataframe to apply OneHotEncoder to the column <code>Orbits</code>, <code>LaunchSite</code>, <code>LandingPad</code>, and <code>Serial</code>. Assign the value to the variable <code>features_one_hot</code>, display the results using the method head. Your result dataframe must include all features including the encoded ones.


```python
# Tạo dummy variables cho các cột phân loại
features_one_hot = pd.get_dummies(
    features, 
    columns=['Orbit', 'LaunchSite', 'LandingPad', 'Serial']
)

# Hiển thị 5 dòng đầu tiên
features_one_hot.head()
```

### TASK  8: Cast all numeric columns to `float64`


Now that our <code>features_one_hot</code> dataframe only contains numbers, cast the entire dataframe to variable type <code>float64</code>


```python
# Ép kiểu toàn bộ dataframe sang float64
features_one_hot = features_one_hot.astype('float64')

# Kiểm tra lại kiểu dữ liệu
features_one_hot.dtypes.head()
```
